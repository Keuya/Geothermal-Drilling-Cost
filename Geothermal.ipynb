```python
import matplotlib.pyplot as plt

# Data for general industry trend
years = [2020, 2021, 2022, 2023, 2024]
general_costs = [10, 9.5, 9, 8.5, 8]

# Data for Fervo Energy (assuming steep decline from 2022 to 2024)
fervo_years = [2020, 2022, 2024]
fervo_costs = [9.4, 7.0, 4.8]

# Data for Integrated Well Construction (converted to USD millions for visualization purposes)
# Assuming a standard depth of 2,500 meters for the calculation
depth = 2500
integrated_costs = [(4584 * depth) / 1e6, (1309 * depth) / 1e6]  # in million USD
integrated_years = [2020, 2024]

plt.figure(figsize=(12, 6))

# Plotting the general industry trend
plt.plot(years, general_costs, label='General Industry Trend', linestyle='-', marker='o')

# Plotting the Fervo Energy trend
plt.plot(fervo_years, fervo_costs, label='Fervo Energy', linestyle='--', marker='x')

# Plotting the Integrated Well Construction trend
plt.plot(integrated_years, integrated_costs, label='Integrated Well Construction (per 2500m)', linestyle='-.', marker='s')

# Adding labels and title
plt.title('Geothermal Drilling Costs (20202024)')
plt.xlabel('Year')
plt.ylabel('Cost per Well (USD Millions)')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
